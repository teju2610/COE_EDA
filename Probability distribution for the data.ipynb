# Probability distribution for the data


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
p=0.6
data=bernoulli.rvs(p,size=1000)
plt.bar([0,1],[sum(data==0),sum(data==1)],color=['red','green'],tick_label=["Failure(0)","Success(1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
lambda_ = 3
x = np.arange(0, 10)
pmf = poisson.pmf(x, lambda_)
plt.bar(x, pmf, color='purple', alpha=0.7)
plt.xlabel("Number of calls per Minute")
plt.ylabel("Probability")
plt.title(f"Poisson Distribution (λ={lambda_})")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu=70
sigma=10
x=np.linspace(mu-4*sigma,mu+4*sigma,100)
pdf=norm.pdf(x,mu,sigma)
plt.plot(x,pdf,color='red',label=f"N({mu},{sigma}**2)")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (mu={mu},sigma={sigma})")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
z=np.linspace(-4,4,1000)
pdf=norm.pdf(z,0,1)
plt.plot(z,pdf,color='blue',label="Standard Normal Distribution (mu=0,sigma=1)")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution(Z-Distribution)")
plt.axvline(x=0,color='red',linestyle='--',label="Mean(z=0)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
income_data=np.random.exponential(scale=50,size=1000)
plt.figure(figsize=(8,4))
sns.histplot(income_data,bins=50,kde=True,color="r")
plt.title("Original Income Distribution (Skewed)")
plt.xlabel("Income($1000)")
plt.ylabel("Frequency")
plt.show()
    
```

```python
sample_means=[]
n_samples=1000
sample_size=30
for _ in range(n_samples):
    sample=np.random.choice(income_data,size=sample_size)
    sample_means.append(np.mean(sample))
plt.figure(figsize=(8,4))
sns.histplot(sample_means,bins=50,kde=True,color="b")
plt.title(f"Sampling Distribution of Sample Mean(n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
from scipy.stats import lognorm

shape = 0.5  
scale = np.exp(10)  
x = np.linspace(0, 200000, 1000)
pdf = lognorm.pdf(x, shape, scale=scale)

plt.figure(figsize=(8,5))
plt.plot(x, pdf, color='blue', label="Log-Normal Distribution")
plt.xlabel("Salary ($)")
plt.ylabel("Probability Density")
plt.title("Log-Normal Distribution of Salaries")
plt.axvline(scale, color='red', linestyle='--', label="Median Salary")
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import powerlaw

np.random.seed(42)
data = np.random.zipf(a=2.5, size=10000)
a, loc, scale = powerlaw.fit(data, floc=0)  
print(f"Estimated alpha: {a}")

plt.figure(figsize=(8,5))
plt.hist(data[data < 50], bins=50, density=True, log=True, alpha=0.7, color='g')
plt.xlabel("Followers Count")
plt.ylabel("Probability Density")
plt.title("Power Law Distribution (Twitter Followers)")
plt.grid()
plt.show()
```
